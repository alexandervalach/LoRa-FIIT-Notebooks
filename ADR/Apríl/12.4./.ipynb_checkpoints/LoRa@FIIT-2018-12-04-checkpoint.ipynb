```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ttn_data = ttn_data[np.isfinite(ttn_data['metadata__frequency'])]
frequencies = ttn_data['metadata__frequency'].unique()
frequencies
```

## Analýza atribútov

### Využitie frekvenčného pásma

```python
plt.hist(ttn_data['metadata__frequency'])
plt.xlabel('Frekvencia [MHz]')
plt.savefig('frequency.png')
```

### Využitie kanálov LoRaWAN základne

```python
plt.hist(ttn_data['metadata__gateways__channel'])
plt.xlabel('Kanály LoRaWAN základne')
plt.savefig('gateways_channel.png')
```

```python
plt.scatter(ttn_data['metadata__gateways__rssi'], ttn_data['metadata__gateways__snr'])
plt.xlabel('RSSI [dBm]')
plt.ylabel('SNR [dB]')
plt.savefig('frequency.png')
```
